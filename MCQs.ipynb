```python
orders = pd.read_csv(datapath + 'orders.csv')
orders
```

```python
orders.info()
```

```python
orders.head()
```

```python
orders = orders.sort_values(['customer_id', 'order_date'])
orders
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['prev_date'] = orders.groupby('customer_id')['order_date'].shift(1)
orders
```

```python
orders['order_gap'] = orders['order_date'] - orders['prev_date']
orders.head()
```

```python
gaps = orders.dropna(subset=['order_gap'])['order_gap']
median_gap = gaps.median()
median_gap
```

**Question 6:**

```python
customers = pd.read_csv(datapath + 'customers.csv')
customers
```

```python
customers = customers[customers['age_group'].notna()]
```

```python
customers_with_orders = customers.merge(orders, on='customer_id')
total_orders = customers_with_orders.groupby('age_group')['order_id'].count()
total_orders
```

```python
num_customers = customers_with_orders.groupby('age_group')['customer_id'].nunique()
num_customers
```

```python
average_orders = total_orders / num_customers
average_orders.sort_values(ascending=False)
```

**Answer:** A. 55+

**Question 9:**

```python
items_with_products = order_items.merge(products[['product_id', 'size']], on='product_id')
num_size = items_with_products['size'].value_counts()
num_size
```

```python
returned = returns.merge(products[['product_id', 'size']], on='product_id')
returned
```

```python
return_quantity = returned.groupby('size')['return_quantity'].sum()
return_quantity
```

```python
num_return_total = return_quantity.to_frame().join(num_size)
num_return_total
```

```python
return_rate = num_return_total['return_quantity'] / num_return_total['count']
return_rate.sort_values(ascending=False)
```
